#MAPS

```python
countries=json.load(open('countries.geo.json'))
```

```python
countries['features'][0].keys()
```

```python
countries['features'][0]['properties']
```

```python
countries_id={}
for feature in countries['features']:
  countries_id[feature['id']]=feature['properties']['name']
  countries_id[feature['properties']['name']]=feature['id']
```

```python
mer['id']=mer['Country'].apply(lambda x:countries_id.get(x,None))
```

#Plotting map of countries vs cpi score 2023

```python
fig = px.choropleth(
    mer,
    locations="id",
    geojson=countries,
    color="CPI score 2023",
    hover_name="Country",
    hover_data=["CPI score 2023"],
    title="World Corruption Rate",
)
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
